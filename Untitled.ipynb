```python
from konlpy.tag import Twitter
import tensorflow as tf
import numpy as np
import codecs
import re

import config
```

```python
twitter = Twitter()
```

```python
class Conversation():
    
    voc_list=[]
    voc_dict=[]
    voc_size=[]
    conversation=[]
    index_in_epoch=0
    
    def Tokenizer(self, sentence):
        token=[]
        for word in sentence.strip().split():
            token.extend(re.compile("([.,!?\"':;)(])").split(word))
        
        ret=[t for t in token if t]
        return ret             
                   

                
    def Load_conversation(self,data_path): # 2 
        self.conversation=[]
        
        with open(data_path,'r',encoding='utf-8') as f:
            for line in f:
                token=twitter.morphs(line.strip())
                index=self.Token_to_index(token)
                print("index" ,index)
                self.conversation.append(index)
        #print(self.conversation)
        
    def Token_to_index(self, token):
        index=[]
        for word in token:
            if word in self.voc_dict:
                index.append(self.voc_dict[word])
            else:
                index.append(config.UNK_ID)
        return index
    
    def Load_voc(self, voc_path): # 1 
        
        front=config.ALL+[]
        self.voc_list=front
        
        with open(voc_path,'r', encoding='utf-8') as f:
            for word in f:
                self.voc_list.append(word.strip())
                
        self.voc_dict={n: i for i, n in enumerate(self.voc_list)}
        self.voc_size= len(self.voc_list)
        
        
        
def main():
    conv=Conversation()
    conv.Load_voc(config.VOC_PATH)
    conv.Load_conversation(config.DATA_PATH)
    
        
if __name__=="__main__":
    main()
    
```
